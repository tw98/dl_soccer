```python
import torch
import torch.nn.functional as F
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import numpy as np
import pandas as pd
from scipy.stats import zscore
```

```python
country_list = ['england', 'germany', 'france', 'italy', 'spain']
dfs = []
for country in country_list:
    # df = (pd.read_csv(f'data/final_dfs/{country}.csv')).drop(columns=['shot', 'corner', 'freeKickShots', 'redCard', 'yellowCard', 'offside'])
    df = (pd.read_csv(f'data/final_dfs/{country}_adv.csv'))
    dfs.append(df)

game_events = pd.concat(dfs, ignore_index=True)
```

```python
dfs_underdogs = []
for country in country_list:
    df = (pd.read_csv(f'data/underdogs/underdogs_{country}.csv'))
    dfs_underdogs.append(df)

underdog_games = pd.concat(dfs_underdogs, ignore_index=True)
underdog_games.shape
```

```python
underdog_games
```

```python
def is_underdog_win(winner, team1, team2, team1_underdog, team2_underdog):
    if (winner == team1 and team1_underdog):
        return True
    elif (winner == team2 and team2_underdog):
        return True
    else:
        return False

underdog_games['underdog_win'] = underdog_games.apply(lambda x: 
                                                      is_underdog_win(x['winner'], x['team1_id'], x['team2_id'], x['HomeUnderdog'], x['AwayUnderdog']), 
                                                      axis='columns')
```

```python
underdog_games[underdog_games['underdog_win'] == True]
```

```python
new_cols = []
for col in list(game_events.columns):
    if 'team1' in col:
        new_cols.append(col.replace('team1', 'underdog'))
    elif 'team2' in col:
        new_cols.append(col.replace('team2', 'favorite'))
    else:
        new_cols.append(col)
        
underdog_games_dict = dict([(x, []) for x in new_cols])
```

```python
for row in underdog_games.iterrows():
    game_id = row[1].wyId
    underdog_win = row[1].underdog_win
    
    if row[1].HomeUnderdog:
        underdog_team = 'team1'
        favorite_team = 'team2'
    else:
        underdog_team = 'team2'
        favorite_team = 'team1'
        
    goi = game_events[game_events['game_id'] == game_id]
    
    for col in list(goi.columns):
        if underdog_team in col:
            target_col = col.replace(underdog_team, 'underdog')
            underdog_games_dict[target_col].append(goi[col].item())
        elif favorite_team in col:
            target_col = col.replace(favorite_team, 'favorite')
            underdog_games_dict[target_col].append(goi[col].item())
        else:
            underdog_games_dict[col].append(goi[col].item())
    
```

```python
df_dataset = pd.DataFrame.from_dict(underdog_games_dict)
```

```python
df_dataset = df_dataset.fillna(0)
```

```python
drop_cols = ['underdog_id', 'underdog_name', 'favorite_id', 'favorite_name', 'underdog_goals', 'favorite_goals', 'favorite_win']
df_dataset = df_dataset.drop(columns=drop_cols)
```

```python
df_dataset.loc[:]['underdog_win'] = df_dataset.apply(lambda x: int(x['underdog_win']), axis = 'columns')
df_dataset = df_dataset.astype('float')
```

```python
# zscore columns of dataframe 
exclude = set(['game_id', 'underdog_win'])
for key in (set(df_dataset.keys()) - exclude):
    if np.std(df_dataset.loc[:][key]) != 0:
        df_dataset.loc[:][key] = zscore(df_dataset[key].astype(float))
```

```python
test_percentage = 0.4

n_games, n_features = dataset[:,2:].shape
trainset, testset = torch.utils.data.random_split(dataset, [n_games - int(test_percentage * n_games), int(test_percentage * n_games)])

train_loader = torch.utils.data.DataLoader(trainset, batch_size=128, shuffle=True)
test_loader = torch.utils.data.DataLoader(testset, batch_size=128, shuffle=True)
```

```python
model = torch.load('./model')
```

```python
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=1e-5)
# criterion = nn.MSELoss()
criterion = nn.BCELoss()
```

```python
def eval(dataset):
    model.eval()
    eval_loader = torch.utils.data.DataLoader(dataset, batch_size=len(dataset),shuffle=False)
    
    batch = next(iter(eval_loader))    
    targets = batch[:,1].float()
    data = batch[:,2:].float()

    truth_team1_win = np.count_nonzero(targets.numpy() == 1.)
    truth_team2_win = np.count_nonzero(targets.numpy() == 0.)

    with torch.no_grad():
        output = model.forward(data)
        pred = torch.round(output.view(-1))

        correct = pred.eq(targets.data.view_as(pred)).sum().item()

        pred_team1_win = np.count_nonzero(pred.numpy() == 1.)
        pred_team2_win = np.count_nonzero(pred.numpy() == 0.)

    model.train()
    
    print('Test Set Accuracy {}/{} ({})'.format(correct, len(dataset), round(correct/len(dataset), 5)))
    # print('Team 1 Win: P: {}, T: {} -- Team 2 Win: P: {}, T: {}'.format(pred_team1_win, truth_team1_win, pred_team2_win, truth_team2_win))

def train():
    for ep in range(1, N_EPOCHS+1):
        model.train()
        print(f' -- Epoch {ep} --')
        loss_sum = 0

        for batch_idx, batch in enumerate(train_loader):
            optimizer.zero_grad()

            targets = batch[:,1].float()
            data = batch[:,2:]

            pred = model.forward(data.float())

            loss = criterion(pred.view(-1), targets)
            
            loss.backward()

            optimizer.step()

        eval(testset)
        
```

```python
print('BEFORE TRAINING')
eval(testset)

print('train...')
train()
```

```python
def get_game_predictions(dataset):
    model.eval()
    eval_loader = torch.utils.data.DataLoader(dataset, batch_size=len(dataset),shuffle=False)
    
    batch = next(iter(eval_loader))
    game_ids = batch[:,0].int().numpy()
    targets = batch[:,1].float()
    data = batch[:,2:].float()

    truth_team1_win = np.count_nonzero(targets.numpy() == 1.)
    truth_team2_win = np.count_nonzero(targets.numpy() == 0.)

    with torch.no_grad():
        output = model.forward(data)
        pred = torch.round(output.view(-1))

        correct = pred.eq(targets.data.view_as(pred)).sum().item()

        pred_team1_win = np.count_nonzero(pred.numpy() == 1.)
        pred_team2_win = np.count_nonzero(pred.numpy() == 0.)
    
    correct = pred.eq(targets.data.view_as(pred))
    
    print('Test Set Accuracy {}/{} ({})'.format(correct.sum().item(), len(dataset), round(correct.sum().item()/len(dataset), 5)))

    correct_pred = [(game, prediction.item()) for game, prediction, truth in zip(game_ids, pred, correct) if truth == True]
    false_pred = [(game, prediction.item()) for game, prediction, truth in zip(game_ids, pred, correct) if truth == False]
    
    return correct_pred, false_pred, pred.numpy(), targets.numpy()

correct_predictions, false_predictions, predictions, ground_truth = get_game_predictions(testset)
```
