```python
import torch
import torch.nn.functional as F
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import numpy as np
import pandas as pd
from scipy.stats import zscore
```

```python
country_list = ['england', 'germany', 'france', 'italy', 'spain']
dfs = []
for country in country_list:
    # df = (pd.read_csv(f'data/final_dfs/{country}.csv')).drop(columns=['shot', 'corner', 'freeKickShots', 'redCard', 'yellowCard', 'offside'])
    df = (pd.read_csv(f'data/final_dfs/{country}_adv.csv'))
    dfs.append(df)

game_events = pd.concat(dfs, ignore_index=True)
```

```python
dfs_underdogs = []
for country in country_list:
    df = (pd.read_csv(f'data/underdogs/underdogs_{country}.csv'))
    dfs_underdogs.append(df)

underdog_games = pd.concat(dfs_underdogs, ignore_index=True)
underdog_games.shape
```

```python
game_events = game_events.fillna(0)
```

```python
# cols = ['team1_win', 'team1_height', 'team2_height', 'team1_shot', 'team1_corner', 'team1_freeKickShots', 'team1_redCard', 'team1_yellowCard', 'team1_offside', 'team2_shot', 'team2_corner', 'team2_freeKickShots', 'team2_redCard', 'team2_yellowCard', 'team2_offside']
# df_dataset = game_events[cols]

drop_cols = ['team1_id', 'team1_name', 'team2_id', 'team2_name', 'team1_goals', 'team2_goals', 'team2_win']
df_dataset = game_events.drop(columns=drop_cols)
```

```python
df_dataset.loc[:]['team1_win'] = df_dataset.apply(lambda x: int(x['team1_win']), axis = 'columns')
df_dataset = df_dataset.astype('float')
```

```python
# zscore columns of dataframe 
exclude = set(['game_id', 'team1_win'])
for key in (set(df_dataset.keys()) - exclude):
    if np.std(df_dataset.loc[:][key]) != 0:
        df_dataset.loc[:][key] = zscore(df_dataset[key].astype(float))
```

```python
dataset = df_dataset.to_numpy()
```

```python
dataset.shape
```

```python
n_games, n_features = dataset[:,2:].shape
trainset, testset = torch.utils.data.random_split(dataset, [n_games - int(0.25 * n_games), int(0.25 * n_games)])

train_loader = torch.utils.data.DataLoader(trainset, batch_size=128, shuffle=True)
test_loader = torch.utils.data.DataLoader(testset, batch_size=128, shuffle=True)
```

```python
n_games
```

```python
N_EPOCHS = 30
BATCH_SIZE_TRAIN = 32
BATCH_SIZE_TEST = 1000
LEARNING_RATE = 0.00001

LOG_TRAINING = False
LOG_INTERVAL = 10

# random_seed = 1
torch.backends.cudnn.enabled = False
# torch.manual_seed(random_seed)
```

```python
# Network Parameters
n_hidden_1 = 1024  # 1st layer number of neurons
n_hidden_2 = 512  # 2nd layer number of neurons
n_hidden_3 = 128  # 3rd layer number of neurons
n_hidden_4 = 64  # 3rd layer number of neurons


model = nn.Sequential(nn.Linear(n_features, n_hidden_1),
                       nn.Tanh(),
#                        nn.Dropout(0.1),
                       nn.Linear(n_hidden_1, n_hidden_2),
                       nn.Tanh(),
#                        nn.Dropout(0.1),
                       nn.Linear(n_hidden_2, n_hidden_3),
                       nn.Tanh(),
                       nn.Linear(n_hidden_3, n_hidden_4),
                       nn.Tanh(),
                       nn.Linear(n_hidden_4, 1),
                       nn.Sigmoid())
```

```python
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=1e-5)
# criterion = nn.MSELoss()
criterion = nn.BCELoss()
```

```python
def eval(dataset, label):
    model.eval()
    eval_loader = torch.utils.data.DataLoader(dataset, batch_size=len(dataset),shuffle=False)
    
    batch = next(iter(eval_loader))    
    targets = batch[:,1].float()
    data = batch[:,2:].float()

    truth_team1_win = np.count_nonzero(targets.numpy() == 1.)
    truth_team2_win = np.count_nonzero(targets.numpy() == 0.)

    with torch.no_grad():
        output = model.forward(data)
        
        loss = criterion(output.view(-1), targets)
        
        pred = torch.round(output.view(-1))

        correct = pred.eq(targets.data.view_as(pred)).sum().item()

        pred_team1_win = np.count_nonzero(pred.numpy() == 1.)
        pred_team2_win = np.count_nonzero(pred.numpy() == 0.)

    model.train()
    
    print('{} Set Accuracy {}/{} ({})'.format(label, correct, len(dataset), round(correct/len(dataset), 5)))
    # print('Team 1 Win: P: {}, T: {} -- Team 2 Win: P: {}, T: {}'.format(pred_team1_win, truth_team1_win, pred_team2_win, truth_team2_win))
    
    return loss, round(correct/len(dataset), 5)

def train():
    train_losses, test_losses = [], []
    train_accuracies, test_accuracies = [], []
    
    for ep in range(1, N_EPOCHS+1):
        model.train()
        print(f' -- Epoch {ep} --')
        loss_sum = 0

        for batch_idx, batch in enumerate(train_loader):
            optimizer.zero_grad()

            targets = batch[:,1].float()
            data = batch[:,2:]

            pred = model.forward(data.float())

            loss = criterion(pred.view(-1), targets)
            
            loss.backward()

            optimizer.step()
            
        train_loss, train_accuracy = eval(trainset, 'Train')
        test_loss, test_accuracy = eval(testset, 'Test')
        
        train_losses.append(train_loss)
        test_losses.append(test_loss)
        
        train_accuracies.append(train_accuracy)
        test_accuracies.append(test_accuracy)
        
    return train_losses, test_losses, train_accuracies, test_accuracies
```

```python
print('BEFORE TRAINING')
eval(testset, 'Test')

print('train...')
train_losses, test_losses, train_accuracies, test_accuracies = train()
```

```python
def get_game_predictions(dataset):
    model.eval()
    eval_loader = torch.utils.data.DataLoader(dataset, batch_size=len(dataset),shuffle=False)
    
    batch = next(iter(eval_loader))
    game_ids = batch[:,0].int().numpy()
    targets = batch[:,1].float()
    data = batch[:,2:].float()

    truth_team1_win = np.count_nonzero(targets.numpy() == 1.)
    truth_team2_win = np.count_nonzero(targets.numpy() == 0.)

    with torch.no_grad():
        output = model.forward(data)
        pred = torch.round(output.view(-1))

        correct = pred.eq(targets.data.view_as(pred)).sum().item()

        pred_team1_win = np.count_nonzero(pred.numpy() == 1.)
        pred_team2_win = np.count_nonzero(pred.numpy() == 0.)
    
    correct = pred.eq(targets.data.view_as(pred))
    correct_pred = [(game, prediction.item()) for game, prediction, truth in zip(game_ids, pred, correct) if truth == True]
    false_pred = [(game, prediction.item()) for game, prediction, truth in zip(game_ids, pred, correct) if truth == False]
    return correct_pred, false_pred, pred.numpy(), targets.numpy()

correct_predictions, false_predictions, preds, truths = get_game_predictions(testset)

```

```python
import matplotlib.pyplot as plt

def visualize_importances(feature_names, importances, y_range, 
                          title="Average Feature Importances", 
                          plot=True, axis_title="Features"):
    print(title)
    for i in range(len(feature_names)):
        print(feature_names[i], ": ", '%.3f'%(importances[i]))
    x_pos = (np.arange(len(feature_names)))
    if plot:
        plt.figure(figsize=(12,6))
        plt.bar(x_pos, importances, align='center')
        plt.xticks(x_pos, feature_names, rotation = 90, wrap=True)
        plt.xlabel(axis_title)
        plt.ylim(y_range)
        plt.title(title)
```
